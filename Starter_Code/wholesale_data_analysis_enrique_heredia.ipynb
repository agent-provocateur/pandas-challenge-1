## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty

df['subtotal'] = df['unit_price'] * df['qty']
df[['client_id', 'first', 'last', 'order_date', 'subcategory', 'unit_price', 'qty', 'subtotal']].head(5)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for orders 50 pounds or under.

df['total_weight'] = df['qty'] * df['unit_weight']

def shipping(val):
    if val <= 50:
        return val * 10
    return val * 7
df['shipping_price'] = df['total_weight'].apply(shipping)
df[['client_id', 'first', 'last', 'order_date', 'subcategory', 'unit_price', 'unit_weight', 'qty', 'total_weight', 'subtotal', 'shipping_price']].head(5)
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%

df['total'] = (df['subtotal'] + df['shipping_price']) * 1.0925
df['total_price'] = round (df['total'], 2)
df[['client_id', 'first', 'last', 'order_date', 'subcategory', 'unit_price', 'unit_weight', 'qty', 'total_weight', 'subtotal', 'shipping_price', 'total_price']].head(5)
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).

df['cost'] = (df['unit_cost'] * df['qty']) + df['shipping_price']
df[['client_id', 'order_date', 'subcategory', 'unit_cost', 'unit_price', 'unit_weight', 'qty', 'total_weight', 'subtotal', 'shipping_price', 'cost', 'total_price']].head(5)
```

```python
# Create a column for the profit of each line using line cost and line price

# NOTE: The 'total_price' includes the tax (in NY is correct to add tax to the shipping cost)
# The 'cost' includes 'shiping' per instructions on line 101
# However, the profit should not include tax nor the 'cost' from previous calculations should be considered
# Therefore a different calculation not includiing 'total_price' or 'cost' is required.

df['cost_before_shipping'] = df['unit_cost'] * df['qty']
df['profit'] = df['subtotal'] - df['cost_before_shipping']
df['profit(%)'] = round((df['profit'] * 100) / (df['cost_before_shipping']))
df['tax'] = (df['subtotal'] + df['shipping_price']) * 0.0925
df['tax'] = round(df['tax'], 2)
df[['client_id', 'unit_cost', 'unit_price', 'qty', 'cost_before_shipping', 'shipping_price', 'subtotal', 'tax', 'total_price', 'profit', 'profit(%)']]
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.

clients = [33615, 66037, 46820, 38378, 24741]
total_spent_top5 = []
counter = -1
for client in clients:
    counter += 1
    top_client = df.loc[df['client_id'] == client]
    total_top5 = round(sum(top_client['total_price']), 2)
    print(f"Client ID {clients[(counter)]} spent ${total_top5:,}")
    total_spent_top5.append(total_top5)  
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 

total_units_top5= []
total_shipping_top5= []
total_revenue_top5= []
total_profit_top5= []
clients = [33615, 66037, 46820, 38378, 24741]
counter = -1

for client in clients:
    counter += 1
    top_client = df.loc[df['client_id'] == client]
    units_top5 = sum(top_client['qty'])
    shipping_top5 = round(sum(top_client['shipping_price']), 2)
    revenue_top5 = round(sum(top_client['total_price']), 2)
    profit_top5 = round(sum(top_client['profit']), 2)
    total_units_top5.append(units_top5) 
    total_shipping_top5.append(shipping_top5)
    total_revenue_top5.append(revenue_top5)
    total_profit_top5.append(profit_top5)
    
summary_data = {"total units purchased" : total_units_top5, 
                "total shipping price" : total_shipping_top5,
                "total revenue" : total_spent_top5,
                "total profit" : total_profit_top5
               }
summary_df = pd.DataFrame(summary_data, clients)
summary_df
```

```python
# Format the data and rename the columns to names suitable for presentation.
# Define the money columns. 
# Define a function that converts a dollar amount to millions.
# Apply the currency_format_millions function to only the money columns. 
# Rename the columns to reflect the change in the money format. 

summary_df_renamed = summary_df
summary_df_renamed.rename(columns = {'total units purchased' : 'Units Purchased',
                           'total shipping price' : 'Shipping Total in Mill $',
                           'total revenue' : 'Total Revenue in Mill $',
                           'total profit' : 'Total Profit in Mill $'},
                          inplace = True)
val=0
def millions(val):
    return val / 1000000
summary_df_renamed['Shipping Total in Mill $'] = summary_df_renamed['Shipping Total in Mill $'].apply(millions)
summary_df_renamed['Total Revenue in Mill $'] = summary_df_renamed['Total Revenue in Mill $'].apply(millions)
summary_df_renamed['Total Profit in Mill $'] = summary_df_renamed['Total Profit in Mill $'].apply(millions)
summary_df_renamed
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.

sorted_summary_df = summary_df.sort_values(by=('Total Profit in Mill $'), ascending=False)
sorted_summary_df
```
